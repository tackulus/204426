```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_csv("/content/insurance.csv")
data.head()
```

# 1. จงเขียนโปรแกรมเพื่อตรวจสอบลักษณะการกระจายของค่าตัวแปร age ว่ามีลักษณะการแจกแจงแบบใด (2 คะแนน)

```python
res = stats.probplot(data['age'], dist=stats.uniform, plot=plt)
```

# 2. จงเขียนโปรแกรมเพื่อแสดงภาพเปรียบเทียบลักษณะการจายของค่าตัวแปร expenses ระหว่างกลุ่มเพศชายและหญิง (2 คะแนน)

```python
sns.boxplot(data=data, x="sex", y='expenses')
```

# 3. จงเขียนโปแกรมเพื่อแสดงลักษณะความสัมพันธ์ระหว่างค่าตัวแปร bmi และ expenses (1 คะแนน)

```python
sns.scatterplot(data=data, x='bmi', y='expenses')
```
